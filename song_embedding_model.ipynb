```python
import pandas as pd
from urllib import request
```

```python
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')
```

```python
lines = data.read().decode("utf-8").split('\n')[2:]
```

```python
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]
```

```python
import pandas as pd
from urllib import request

# Load file
songs_file = request.urlopen(
    'https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt'
)
songs_file = songs_file.read().decode("utf-8").strip().split('\n')

# Split rows and keep only valid rows with 3 columns
songs = [s.rstrip().split('\t') for s in songs_file if len(s.split('\t')) == 3]

# Create DataFrame
songs_df = pd.DataFrame(songs, columns=['id', 'title', 'artist'])

# Set index
songs_df = songs_df.set_index('id')

songs_df.head()
```

```python
print( 'Playlist #1:\n ', playlists[0], '\n')
print( 'Playlist #2:\n ', playlists[1])
```

```python
from gensim.models import Word2Vec

model = Word2Vec(playlists, vector_size=32, window=20, negative=50,
min_count=1, workers=4)
```
